## Building a Classification Model for the iris dataset

In this notebook we will build a classification model for the iris dataset using the random forest algorithm

###1. Import libraries and load the iris dataset


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

iris = datasets.load_iris()
iris      #calling iris
```

##2. Input and Output features
We can see that the iris dataset has 4 input features and 1 output variable which is the call type



```python
iris.feature_names # explicitly printing the features
```

```python
print(iris.target_names) #explicitly printing the target names
```

```python
#we can also print the target name numerically, represented by 0 1 and 2
print(iris.target)   # this was you can see the number of each target (50 each)
```

##3. A Glimpse of the data

```python
print(iris.data) # will print an array of the dataset
```

###4.1 Assigning input and output variables and examining the data dimensions

Lets assign the 4 features to X and the label will be Y. With a total of 150 data samples which represent 150 flowers that belong to 1 of 3 classes.

```python
x= iris.data #contains four input features
y = iris.target #contain 1 variable

#examining the data dimensions
print(f"The x shape: {x.shape}")
print(f"The y shape: {y.shape}")
```

`(150, 4)` tells us that there are 150 rows and 4 columns which correspond to 150 flowers and 4 features.
`(150,)` tells us only the number of rows (the flowers) and nothing after because it is one column.

##5. Build the Classification Model using Random Forest

```python
clf = RandomForestClassifier()
#call the RFC
clf.fit(x, y)     #this calls the RFC and using the fit function to create the classification model that take x and y as its 2 arguments
#it will take x (features) and create a model by using y as the class label
# so the model learns in a supervised manner by learning from the class label and outputing a classification model using RFC
```

###6. Feature Importance

```python
print(clf.feature_importances_)
```

The above results show the difference importance each difference feature has to the classification model.  `0.10603183` corresponds to sepal lenth; `0.02360766` corresponds to sepal width; `0.43168005` corresponds petal width; and `0.43868046` correspondds to petal lenth  

###7. Make Predictions

```python
#we can check the values of the first flower by
print(x[0])
```

```python
# lets check if out prict will work using the values of the x[0]
print(clf.predict([[5.1, 3.5, 1.4, .2]]))

#OR by calling the array index
print(clf.predict([x[0]]))
```

The output is [0] which is the numerical representation of the target name setosa class

```python
#the following function will tell us the probability of the prediction, is it class 1 (at what chances) or is it class 2 (eg. at 0.2) or...
print(clf.predict_proba([x[0]]))
```

This tells us that the prediction is a 100% sure that the first class x[0] is setora and that there is no probability that it is any other clss type

```python
#This will change the outputs from being numerically represented to the actual class name
clf.fit(iris.data, iris.target_names[iris.target])
```

```python
print(clf.predict([x[0]])) #instead of [0] it will now print the name
```

##8. Data Split (80/20 ratio)


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2) #this take the input argument x y variables and the test ratio means .8 will be for the train
#then lets show the shape of the train and test shape
print(f"This is the training shape of x:{x_train.shape} and y: {y_train.shape}")

print(f"This is the testing shape: {x_test.shape, y_test.shape}")
```

##9. Building the Random Forest Model

```python
clf.fit(x_train, y_train)
```

###9.1 Preform predictions on the single sample from the dataset

```python

print(clf.predict([x[0]]))
```

```python
print(clf.predict_proba([x[0]]))
```

###9.2 Perform prediction on the test set

```python

#if we what to predict a set of samples by used the whole x_test which contain 30 flowers
print(clf.predict(x_test))
```
